```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
v = school_data_complete
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
school_count = v['school_name'].nunique()
student_count = v['Student ID'].nunique()
total_budget = school_data['budget'].sum()
total_budget = '${:,.2f}'.format(total_budget)
avg_mathscore = v['math_score'].mean()
avg_readingscore = v['reading_score'].mean()
ovr_passingrate = (avg_mathscore + avg_readingscore) / 2
math_70    = (len(v[v['math_score'] > 69]) / student_count) * 100
reading_70 = (len(v[v['reading_score'] > 69]) / student_count) * 100

summary_df = pd.DataFrame({"School Count": [school_count],
                           "Student Count": [student_count],
                           "Total Budget": [total_budget],
                           "Avg Math Score": [avg_mathscore],
                           "Avg Reading Score": [avg_readingscore],
                           "Overall Passing Percentage": [ovr_passingrate],
                           "% Passing Math": [math_70],
                           "% Passing Reading": [reading_70],
                           "% Overall Passing Rate": [ovr_passingrate]})

summary_df.head()
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
school_district = [{'school_name': 'Huang High School', 'type': 'District'},
                  {'school_name':'Figueroa High School', 'type':'District'},
                  {'school_name':'Shelton High School', 'type':'Charter'},
                  {'school_name':'Hernandez High School', 'type':'District'},
                  {'school_name':'Griffin High School', 'type':'Charter'},
                  {'school_name':'Wilson High School', 'type':'Charter'},
                  {'school_name':'Cabrera High School', 'type':'Charter'},
                  {'school_name':'Bailey High School', 'type':'District'},
                  {'school_name':'Holden High School', 'type':'Charter'},
                  {'school_name':'Pena High School', 'type':'Charter'},
                  {'school_name':'Wright High School', 'type':'Charter'},
                  {'school_name':'Rodriguez High School', 'type':'District'},
                  {'school_name':'Johnson High School', 'type':'District'},
                  {'school_name':'Ford High School', 'type':'District'},
                  {'school_name':'Thomas High School', 'type':'Charter'}]
  

grouped = v.groupby('school_name')

num_students = grouped['Student ID'].count()
school_type = grouped['type'].unique()
total_budget_by_school = grouped['budget'].max()
per_student_budget = total_budget_by_school / num_students
avg_math_per_student = grouped['math_score'].mean()
avg_reading_per_student = grouped['reading_score'].mean()
                   
#mathabove69
ismathabove69 = v['math_score'] > 69
mathabove69 = v[ismathabove69]
groupedmath = mathabove69.groupby(['school_name'])
num_students_math1 = groupedmath['school_name'].count()
num_students_math2 = grouped['school_name'].count()
final_math = num_students_math1 / num_students_math2 * 100

#readingabove69
isreadingabove69 = v['reading_score'] > 69
readingabove69 = v[isreadingabove69]
groupedreading = readingabove69.groupby(['school_name'])
num_students_reading1 = groupedreading['school_name'].count()
num_students_reading2 = grouped['school_name'].count()
final_reading = num_students_reading1 / num_students_reading2 * 100

avg_ovr_passing = (final_reading + final_math) / 2


summary_table = pd.DataFrame({"Number of Students": num_students,
                              "Total Budget": total_budget_by_school,
                              "Budget per Student": per_student_budget,
                              "Avg Math Score": avg_math_per_student,
                              "Avg Reading Score": avg_reading_per_student,
                              "% Passing Math": final_math,
                              "% Passing Reading": final_reading,
                              "Overall Passing Rate": avg_ovr_passing})

school_district_df = pd.DataFrame.from_dict(school_district)

final_table_staging = pd.merge(summary_table, school_district_df, how="left", on=["school_name", "school_name"])

final_table = final_table_staging.sort_values("Overall Passing Rate", ascending=False)
final_table = final_table.reindex_axis(['school_name','type','Number of Students','Total Budget','Budget per Student','Avg Math Score','Avg Reading Score','% Passing Math','% Passing Reading','Overall Passing Rate'], axis=1)
final_table.reset_index()
final_table.head()
```
